```python
# Load GPT-2 model for text generation
print("Loading AI model... (this may take a minute)")
text_generator = pipeline('text-generation', model='gpt2')
print("✅ AI model loaded!")

def generate_meme_caption(topic, style="funny"):
    """Generate meme caption based on topic"""

    # Create different prompts based on style
    prompts = {
        "funny": f"A hilarious meme about {topic}. Top text:",
        "relatable": f"A relatable meme about {topic}. Caption:",
        "sarcastic": f"A sarcastic meme about {topic}. Text:",
    }

    prompt = prompts.get(style, prompts["funny"])

    # Generate text
    result = text_generator(prompt, max_length=30, num_return_sequences=1,
                           temperature=0.9, do_sample=True)

    # Clean the output
    caption = result[0]['generated_text']
    caption = caption.replace(prompt, '').strip()

    # Limit length
    if len(caption) > 50:
        caption = caption[:50] + "..."

    return caption

# Test the function
test_caption = generate_meme_caption("Monday morning")
print(f"\nTest caption: {test_caption}")
```

```python
# Install additional models for better captions
!pip install openai anthropic -q  # Optional: for GPT-3.5/Claude (requires API key)

from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

class AdvancedCaptionGenerator:
    def __init__(self):
        print("🤖 Loading multiple AI models...")

        # Model 1: GPT-2 (fast, good for quick generations)
        self.gpt2_generator = pipeline('text-generation', model='gpt2')

        # Model 2: DistilGPT2 (faster, lighter)
        self.distilgpt2_generator = pipeline('text-generation', model='distilgpt2')

        # Model 3: GPT-2 Medium (better quality, slower)
        try:
            self.gpt2_medium = pipeline('text-generation', model='gpt2-medium')
            print("✅ GPT-2 Medium loaded!")
        except:
            print("⚠️ GPT-2 Medium failed, using standard GPT-2")
            self.gpt2_medium = self.gpt2_generator

        print("✅ All models loaded!")

        # Meme styles and their prompts
        self.meme_styles = {
            'sarcastic': [
                "Oh great, {topic}. Because that's exactly what I needed.",
                "When {topic} happens: *pretends to be shocked*",
                "{topic}? More like {topic} for dummies."
            ],
            'relatable': [
                "Me trying to {topic}",
                "Everyone: normal life | Me: {topic}",
                "Nobody: | Absolutely nobody: | Me: {topic}"
            ],
            'funny': [
                "POV: You just {topic}",
                "{topic} be like:",
                "When the {topic} hits different"
            ],
            'wholesome': [
                "Appreciating {topic} today",
                "{topic} but make it wholesome",
                "Small reminder about {topic}"
            ],
            'chaos': [
                "{topic} at 3 AM be like",
                "My brain during {topic}:",
                "{topic} energy"
            ]
        }

    def generate_caption(self, topic, style='funny', model='gpt2', temperature=0.9):
        """
        Generate meme caption with multiple options

        Args:
            topic: Meme topic
            style: Caption style (sarcastic, relatable, funny, wholesome, chaos)
            model: Which model to use (gpt2, distilgpt2, gpt2-medium)
            temperature: Creativity level (0.7-1.2)
        """

        # Select generator
        if model == 'gpt2-medium':
            generator = self.gpt2_medium
        elif model == 'distilgpt2':
            generator = self.distilgpt2_generator
        else:
            generator = self.gpt2_generator

        # Get style templates
        templates = self.meme_styles.get(style, self.meme_styles['funny'])
        template = random.choice(templates)

        # Create prompt
        prompt = template.format(topic=topic)

        # Generate with better parameters
        results = generator(
            prompt,
            max_length=50,
            num_return_sequences=3,  # Generate 3 options
            temperature=temperature,
            top_k=50,
            top_p=0.95,
            do_sample=True,
            pad_token_id=50256
        )

        # Clean and rank results
        captions = []
        for result in results:
            caption = result['generated_text']
            # Remove prompt
            caption = caption.replace(prompt, '').strip()
            # Clean up
            caption = caption.split('\n')[0]  # Take first line
            caption = caption[:100]  # Limit length
            if len(caption) > 10:  # Only keep reasonable captions
                captions.append(caption)

        # Return best caption (or first if filtering removed all)
        return captions[0] if captions else prompt

    def generate_multiple_captions(self, topic, style='funny', count=5):
        """Generate multiple caption options for user to choose"""
        captions = []

        # Use different models and temperatures for variety
        configs = [
            ('gpt2', 0.8),
            ('gpt2', 1.0),
            ('distilgpt2', 0.9),
            ('gpt2-medium', 0.85),
            ('gpt2', 1.1)
        ]

        for model, temp in configs[:count]:
            caption = self.generate_caption(topic, style, model, temp)
            captions.append(caption)

        return captions

# Initialize advanced generator
print("🚀 Initializing Advanced Caption Generator...")
advanced_generator = AdvancedCaptionGenerator()
print("✅ Ready!")

# Test it
test_captions = advanced_generator.generate_multiple_captions("Monday morning", style='sarcastic', count=3)
print("\n📝 Sample captions:")
for i, caption in enumerate(test_captions, 1):
    print(f"{i}. {caption}")
```

```python
class VideoMemeGenerator:
    def __init__(self):
        self.supported_formats = ['.mp4', '.avi', '.mov', '.gif']

    def add_text_to_video(self, video_path, top_text="", bottom_text="",
                         output_path="output_video_meme.mp4"):
        """Add text overlay to video"""

        print(f"🎬 Processing video: {video_path}")

        # Load video
        video = VideoFileClip(video_path)

        # Video dimensions
        w, h = video.size

        # Create text clips
        clips = [video]

        if top_text:
            txt_top = TextClip(
                top_text,
                fontsize=int(h/12),
                color='white',
                font='Arial-Bold',
                stroke_color='black',
                stroke_width=2,
                method='caption',
                size=(w*0.9, None)
            )
            txt_top = txt_top.set_position(('center', h*0.05)).set_duration(video.duration)
            clips.append(txt_top)

        if bottom_text:
            txt_bottom = TextClip(
                bottom_text,
                fontsize=int(h/12),
                color='white',
                font='Arial-Bold',
                stroke_color='black',
                stroke_width=2,
                method='caption',
                size=(w*0.9, None)
            )
            txt_bottom = txt_bottom.set_position(('center', h*0.85)).set_duration(video.duration)
            clips.append(txt_bottom)

        # Composite video with text
        final_video = CompositeVideoClip(clips)

        # Write output
        print("💾 Saving video meme...")
        final_video.write_videofile(
            output_path,
            codec='libx264',
            audio_codec='aac',
            fps=24,
            verbose=False,
            logger=None
        )

        # Clean up
        video.close()
        final_video.close()

        print(f"✅ Video meme saved: {output_path}")
        return output_path

    def create_gif_from_video(self, video_path, output_path="output.gif",
                             duration=None, fps=10):
        """Convert video to GIF"""
        video = VideoFileClip(video_path)

        if duration:
            video = video.subclip(0, min(duration, video.duration))

        video.write_gif(output_path, fps=fps)
        video.close()

        return output_path

    def add_caption_to_gif(self, gif_path, top_text="", bottom_text="",
                          output_path="output_gif_meme.gif"):
        """Add text to existing GIF"""

        # Read GIF
        reader = imageio.get_reader(gif_path)
        fps = reader.get_meta_data()['fps']

        # Process frames
        frames = []
        for frame in reader:
            # Convert to PIL Image
            img = Image.fromarray(frame)
            draw = ImageDraw.Draw(img)

            # Add text (similar to image meme)
            w, h = img.size
            try:
                font = ImageFont.truetype("Impact.ttf", int(h/12))
            except:
                font = ImageFont.load_default()

            # Add top text
            if top_text:
                bbox = draw.textbbox((0, 0), top_text, font=font)
                text_w = bbox[2] - bbox[0]
                x = (w - text_w) / 2
                y = 10
                # Outline
                for adj in range(-2, 3):
                    for adj2 in range(-2, 3):
                        draw.text((x+adj, y+adj2), top_text, font=font, fill='black')
                draw.text((x, y), top_text, font=font, fill='white')

            # Add bottom text
            if bottom_text:
                bbox = draw.textbbox((0, 0), bottom_text, font=font)
                text_w = bbox[2] - bbox[0]
                text_h = bbox[3] - bbox[1]
                x = (w - text_w) / 2
                y = h - text_h - 20
                for adj in range(-2, 3):
                    for adj2 in range(-2, 3):
                        draw.text((x+adj, y+adj2), bottom_text, font=font, fill='black')
                draw.text((x, y), bottom_text, font=font, fill='white')

            frames.append(np.array(img))

        # Save as GIF
        imageio.mimsave(output_path, frames, fps=fps)

        return output_path

# Initialize video generator
video_generator = VideoMemeGenerator()
print("✅ Video Meme Generator ready!")
```

```python
def add_text_to_meme(image_path, top_text="", bottom_text="", output_path="output_meme.jpg"):
    """Add text overlay to meme image"""

    # Open the image
    img = Image.open(image_path)
    draw = ImageDraw.Draw(img)

    # Get image dimensions
    img_width, img_height = img.size

    # Try to use Impact font (classic meme font), fallback to default
    try:
        # Download Impact font
        font_size = int(img_height / 12)
        font_url = "https://github.com/matomo-org/travis-scripts/raw/master/fonts/Impact.ttf"
        font_response = requests.get(font_url)

        with open("Impact.ttf", "wb") as f:
            f.write(font_response.content)

        font = ImageFont.truetype("Impact.ttf", font_size)
    except:
        font = ImageFont.load_default()

    # Function to draw text with outline
    def draw_text_with_outline(text, position):
        x, y = position
        # Draw black outline
        for adj_x in range(-2, 3):
            for adj_y in range(-2, 3):
                draw.text((x + adj_x, y + adj_y), text, font=font, fill='black')
        # Draw white text
        draw.text((x, y), text, font=font, fill='white')

    # Add top text (centered)
    if top_text:
        # Calculate text size and position
        bbox = draw.textbbox((0, 0), top_text, font=font)
        text_width = bbox[2] - bbox[0]
        x = (img_width - text_width) / 2
        y = 10
        draw_text_with_outline(top_text, (x, y))

    # Add bottom text (centered)
    if bottom_text:
        bbox = draw.textbbox((0, 0), bottom_text, font=font)
        text_width = bbox[2] - bbox[0]
        text_height = bbox[3] - bbox[1]
        x = (img_width - text_width) / 2
        y = img_height - text_height - 20
        draw_text_with_outline(bottom_text, (x, y))

    # Save the image
    img.save(output_path)
    return output_path

print("✅ Text overlay function created!")
```

```python
def create_meme(topic, template_name, use_ai=True, custom_top_text="", custom_bottom_text=""):
    """
    Main function to create meme

    Args:
        topic: The meme topic
        template_name: Name of template to use
        use_ai: Whether to use AI to generate text or use custom text
        custom_top_text: Custom top text (if not using AI)
        custom_bottom_text: Custom bottom text (if not using AI)
    """

    # Select template
    template_path = f'meme_templates/{template_name}.jpg'

    # Generate or use custom text
    if use_ai and topic:
        # Generate caption with AI
        caption = generate_meme_caption(topic)

        # Split caption into top and bottom (simple approach)
        words = caption.split()
        mid = len(words) // 2
        top_text = ' '.join(words[:mid])
        bottom_text = ' '.join(words[mid:])
    else:
        top_text = custom_top_text
        bottom_text = custom_bottom_text

    # Create output filename
    output_path = f"meme_{random.randint(1000, 9999)}.jpg"

    # Add text to image
    result_path = add_text_to_meme(template_path, top_text, bottom_text, output_path)

    return result_path

print("✅ Meme generator function created!")
```

```python
def generate_advanced_meme(topic, template, mode, top_text, bottom_text,
                          style, num_captions, media_type, video_file):
    """
    Advanced meme generator with all features

    Args:
        media_type: 'image', 'video', or 'gif'
        video_file: Uploaded video file
    """

    try:
        # IMAGE MEME
        if media_type == "Image":
            use_ai = (mode == "AI Generated")

            if use_ai and not topic:
                return None, "⚠️ Please enter a topic!"

            if not use_ai and (not top_text or not bottom_text):
                return None, "⚠️ Please enter both texts for manual mode!"

            # Generate AI captions
            if use_ai:
                if num_captions > 1:
                    # Generate multiple options
                    captions = advanced_generator.generate_multiple_captions(
                        topic, style=style, count=num_captions
                    )
                    caption_text = captions[0]  # Use first for now
                    options_text = "\n\n🎯 Generated options:\n" + "\n".join(
                        f"{i+1}. {c}" for i, c in enumerate(captions)
                    )
                else:
                    caption_text = advanced_generator.generate_caption(topic, style=style)
                    options_text = ""

                # Split into top/bottom
                words = caption_text.split()
                mid = len(words) // 2
                top_text = ' '.join(words[:mid])
                bottom_text = ' '.join(words[mid:])

            # Create meme
            template_path = f'meme_templates/{template}.jpg'
            output_path = f"meme_{random.randint(1000, 9999)}.jpg"

            add_text_to_meme(template_path, top_text, bottom_text, output_path)

            status = f"✅ Meme generated!\n\nTop: {top_text}\nBottom: {bottom_text}"
            if use_ai and num_captions > 1:
                status += options_text

            return output_path, status

        # VIDEO MEME
        elif media_type == "Video" and video_file is not None:
            use_ai = (mode == "AI Generated")

            # Generate or use manual text
            if use_ai and topic:
                caption_text = advanced_generator.generate_caption(topic, style=style)
                words = caption_text.split()
                mid = len(words) // 2
                top_text = ' '.join(words[:mid])
                bottom_text = ' '.join(words[mid:])

            # Process video
            output_path = video_generator.add_text_to_video(
                video_file.name,
                top_text,
                bottom_text,
                f"video_meme_{random.randint(1000, 9999)}.mp4"
            )

            return output_path, f"✅ Video meme created!\n\nTop: {top_text}\nBottom: {bottom_text}"

        # GIF MEME
        elif media_type == "GIF" and video_file is not None:
            use_ai = (mode == "AI Generated")

            if use_ai and topic:
                caption_text = advanced_generator.generate_caption(topic, style=style)
                words = caption_text.split()
                mid = len(words) // 2
                top_text = ' '.join(words[:mid])
                bottom_text = ' '.join(words[mid:])

            # Check if uploaded file is already GIF or needs conversion
            if video_file.name.endswith('.gif'):
                output_path = video_generator.add_caption_to_gif(
                    video_file.name,
                    top_text,
                    bottom_text,
                    f"gif_meme_{random.randint(1000, 9999)}.gif"
                )
            else:
                # Convert video to GIF first, then add text
                temp_gif = f"temp_{random.randint(1000, 9999)}.gif"
                video_generator.create_gif_from_video(video_file.name, temp_gif, duration=5)
                output_path = video_generator.add_caption_to_gif(
                    temp_gif,
                    top_text,
                    bottom_text,
                    f"gif_meme_{random.randint(1000, 9999)}.gif"
                )

            return output_path, f"✅ GIF meme created!\n\nTop: {top_text}\nBottom: {bottom_text}"

        else:
            return None, "⚠️ Please upload a video/GIF file for video/GIF mode!"

    except Exception as e:
        return None, f"❌ Error: {str(e)}"

# Create advanced Gradio interface
with gr.Blocks(title="Advanced AI Meme Generator 🚀", theme=gr.themes.Soft()) as demo:
    gr.Markdown("""
    # 🎭 Advanced AI Meme Generator
    ### Generate Image, Video, and GIF memes with AI! 🚀
    """)

    with gr.Row():
        with gr.Column(scale=1):
            # Media type selector
            media_type = gr.Radio(
                choices=["Image", "Video", "GIF"],
                value="Image",
                label="📱 Media Type"
            )

            # Mode selector
            mode = gr.Radio(
                choices=["AI Generated", "Manual Text"],
                value="AI Generated",
                label="🤖 Generation Mode"
            )

            # Topic input
            topic = gr.Textbox(
                label="💡 Topic (for AI mode)",
                placeholder="e.g., Monday morning, coding bugs, college life...",
                lines=2
            )

            # AI Style selector
            with gr.Accordion("🎨 AI Caption Style", open=True):
                style = gr.Radio(
                    choices=["funny", "sarcastic", "relatable", "wholesome", "chaos"],
                    value="funny",
                    label="Style"
                )
                num_captions = gr.Slider(
                    minimum=1,
                    maximum=5,
                    value=1,
                    step=1,
                    label="Number of caption options to generate"
                )

            # Template selector (for images)
            template = gr.Dropdown(
                choices=sorted(list(meme_urls.keys())),
                value='drake',
                label="🖼️ Meme Template (for images)",
                interactive=True
            )

            # Video/GIF upload
            video_file = gr.File(
                label="📹 Upload Video/GIF (for Video/GIF mode)",
                file_types=[".mp4", ".avi", ".mov", ".gif"]
            )

            # Manual text inputs
            with gr.Accordion("✍️ Manual Text (for Manual mode)", open=False):
                top_text = gr.Textbox(label="Top Text", placeholder="Enter top text...")
                bottom_text = gr.Textbox(label="Bottom Text", placeholder="Enter bottom text...")

            # Generate button
            generate_btn = gr.Button("🚀 Generate Meme", variant="primary", size="lg")

        with gr.Column(scale=1):
            # Output
            output_media = gr.File(label="Generated Meme", interactive=False)
            status_text = gr.Textbox(label="Status", lines=6)

    # Examples section
    gr.Markdown("### 💡 Quick Examples")
    with gr.Row():
        gr.Examples(
            examples=[
                ["Monday morning", "drake", "AI Generated", "", "", "sarcastic", 1, "Image", None],
                ["College exams stress", "disaster_girl", "AI Generated", "", "", "funny", 1, "Image", None],
                ["Debugging at 3 AM", "this_is_fine", "AI Generated", "", "", "chaos", 1, "Image", None],
                ["Group project", "woman_yelling_at_cat", "AI Generated", "", "", "relatable", 1, "Image", None],
            ],
            inputs=[topic, template, mode, top_text, bottom_text, style, num_captions, media_type, video_file]
        )

    # Info section
    gr.Markdown("""
    ---
    ### 📚 Features:
    - ✅ 50+ Popular meme templates
    - ✅ AI-powered caption generation with multiple styles
    - ✅ Video meme support (.mp4, .avi, .mov)
    - ✅ GIF meme support
    - ✅ Multiple AI caption options
    - ✅ Manual text mode for custom captions

    ### 🎯 Tips:
    - For **AI mode**: Enter a topic and choose a style
    - For **Video/GIF**: Upload your file first
    - Try different styles for varied humor!
    """)

    # Connect button
    generate_btn.click(
        fn=generate_advanced_meme,
        inputs=[topic, template, mode, top_text, bottom_text, style, num_captions, media_type, video_file],
        outputs=[output_media, status_text]
    )

print("✅ Advanced interface created! Launching...")
demo.launch(share=True, debug=True)
```
